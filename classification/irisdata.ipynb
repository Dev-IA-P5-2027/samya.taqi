```python
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pandas as pd 
```

```python
iris=load_iris()
data=pd.DataFrame(iris.data, columns=iris.feature_names)
data["target"]=iris.target
data["target"]=data["target"].map({0: "setosa", 1:"versicolor", 2:"virginica"})
data
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=data, x="target", palette="viridis")
plt.title("Nombre d'échantillons par calasse cible ")
plt.xlabel("Classe cible")
plt.ylabel("Nombre d'échantillons")
plt.show
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(data.drop("target", axis=1))
pca_data=pd.DataFrame(x_pca, columns=["PCA1","PCA2"])
pca_data["traget"]=data["target"]
pca
```

```python
pca_data
```

```python
plt.figure(figsize=(8,6))
pca_data = pca_data.rename(columns={"traget": "target"})
sns.scatterplot(
    data=pca_data,
    x="PCA1",
    y="PCA2",
    hue="target",
    palette="viridis",
    s=100
)
plt.title("Réparation des classes sous forme de clustering(PCA)")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.legend(title='Classe cible')
plt.show()
```

```python
plt.figure(figsize=(8,6))
correlation_matrix= data.drop("target", axis=1).corr()
sns.heatmap(correlation_matrix, annot=True , cmap="coolwarm", square=True)
plt.title("Matrice de corrélation des caractéristiques")
plt.show()
```
