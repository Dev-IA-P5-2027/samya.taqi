```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_excel("data\data dictionary.xlsx")
df
```

```python
df=pd.read_csv(r"data\train.csv", sep=",", low_memory=False)
df
```

```python
# Extraire la marque
df["brand"] = df["Name"].str.split().str[0]
df
```

```python
# Extraire le modèle (1 ou 2 mots après la marque)
df["model"] = df["Name"].str.split().str[1:3].str.join(" ")
df
```

```python
# Extraire le variant (le reste)
df["variant"] = df["Name"].str.split().str[3:].str.join(" ")
df
```

```python
data=df.drop(["Name","variant","New_Price"], axis=1).copy()
data
```

```python
data.rename(columns={"Price": "target"}, inplace=True)
data
```

```python
data["Power"] = data["Power"].str.split().str[0]
data["Engine"] = data["Engine"].str.split().str[0]
data["Mileage"] = data["Mileage"].str.split().str[0]
data
```

```python
price_inr_to_euro=0.009267
data["Price"]=data["Price"]*price_inr_to_euro
data
```

# les colonnes qui contiennet des valeurs manquentes 

```python
data["Mileage"] = pd.to_numeric(data["Mileage"], errors='coerce')
data["Engine "] = pd.to_numeric(data["Engine"], errors='coerce')
data["Power"] = pd.to_numeric(data["Mileage"], errors='coerce')
data
```

```python
plt.figure()
plt.scatter(data["Transmission"], data["Engine"],data["Year"],c=data["Year"] , cmap="BrBG" )
plt.title("distribution deTransmission et Engine en fonction de la cible")
plt.xlabel("Transmission")
plt.ylabel("Engine")
plt.legend()
plt.show()
```

**Encodage**

```python
data["Location"]=pd.factorize(data["Location"])[0]
data["Fuel_Type"]=pd.factorize(data["Fuel_Type"])[0]
data["Transmission"]=pd.factorize(data["Transmission"])[0]
data["Owner_Type"]=pd.factorize(data["Owner_Type"])[0]
data["brand"]=pd.factorize(data["brand"])[0]
data["model"]=pd.factorize(data["model"])[0]
data
```
