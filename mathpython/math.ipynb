# dérivation

```python
from sympy import symbols, Derivative

x = symbols('x')
expr = x**2 + 3*x + 2
 
# Calcul de la dérivée de l'expression par rapport à x
derivative_expr = Derivative(expr, x)
 
# Affichage de la dérivée
print(derivative_expr)
```

```python
#la loi normale 
import scipy.stats as  stats 
from scipy.stats import norm

moyenne=178
sigma=10
valeur= 3
X=stats.norm(moyenne, sigma)
proba= X.cdf(moyenne+valeur)-X.cdf(moyenne-valeur)
print(proba)
```

```python
moyenne=175
segma=10
p=0.2358228

x=norm.ppf(p, loc=moyenne,scale=segma)
print(x)
```

# la taille des hommes

```python
import scipy.stats as stats
from scipy.stats import norm 

mu=179
segma= 10
hauteur_portique= 203

p=stats.norm.cdf(hauteur_portique, mu, segma)

print(p)

```

```python
mu=1.4
segma=0.2
poids_poulet=1.7

probabilté=stats.norm.cdf(poids_poulet,mu, segma)
print(1-probabilté)
```
