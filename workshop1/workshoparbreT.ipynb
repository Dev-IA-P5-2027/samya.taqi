**question:Calculer les statistiques (moyenne et écart-type) des quatre variables explicatives : longueur de sépale, largueur de sépale, longueur de pétale et largeur de pétale.**

```python
#pour le calcul statistiques 
import scipy  #scipy une bibliothéque open-source dédiée au calcul scientifique 
scipy.stats.describe(iris.data[:,:5])
```

```python
from tabulate import tabulate
""" au lieu d'afficher à la sortie un tuple je veux afficher un tableau plus clair """
res=scipy.stats.describe(iris.data[:,:5])
table = [
    ["Min"] + list(res.minmax[0]),
    ["Max"] + list(res.minmax[1]),
    ["Moyenne"] + list(res.mean),
    ["Variance"] + list(res.variance),
    ["Skewness"] + list(res.skewness),
    ["Kurtosis"] + list(res.kurtosis)
]  #
headers = ["Statistique"] + iris.feature_names

print(tabulate(table, headers=headers, tablefmt="grid"))
#tablefmt définir le style de mise en forme du tableau et grid tableau avec des bordures complètes 
```

**Refaites une division apprentissage/test avec 5% des données en apprentissage et 95% test.Calculez le taux d’éléments mal classifiés sur l’ensemble de test.Faites varier (ou mieux, réalisez une recherche par grille avec GridSearchCV) les valeurs des paramètres max_depth et min_samples_leaf pour mesurer leur impact sur ce score.**

```python
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(iris.data, iris.target,
            test_size=0.95, random_state=0)

for mdepth in [1, 2, 3, 4, 5, 6, 7]:
    clf = tree.DecisionTreeClassifier(max_depth=mdepth)
    clf = clf.fit(X_train, y_train)
    print(clf.score(X_test, y_test))

for msplit in [2, 3, 5, 10, 15, 20]:
    clf = tree.DecisionTreeClassifier(min_samples_split=msplit)
    clf = clf.fit(X_train, y_train)
    print(clf.score(X_test, y_test))
```

```python
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(iris.data, iris.target,
    test_size=0.30, random_state=0)
from sklearn.model_selection import GridSearchCV
pgrid = {"max_depth": [1, 2, 3, 4, 5, 6, 7],
      "min_samples_split": [2, 3, 5, 10, 15, 20]}
grid_search = GridSearchCV(tree.DecisionTreeClassifier(), param_grid=pgrid, cv=10)
#chercher de bonnes valeurs pour les paramètres on peut utiliser la validation croisée avec gridserach
grid_search.fit(X_train, y_train)
grid_search.best_estimator_.score(X_test, y_test)
```

**Affichage de la surface de décision**

```python
# Paramètres
n_classes = 3
plot_colors = "bry" # blue-red-yellow
plot_step = 0.02

# Choisir les attributs longueur et largeur des pétales
pair = [2, 3]

# On ne garde seulement les deux attributs
X = iris.data[:, pair]
y = iris.target

# Apprentissage de l'arbre
clf = tree.DecisionTreeClassifier().fit(X, y)

# Affichage de la surface de décision
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step)) #meshgrid gérer le calcul numérique et les tableaux 
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired) 
#trace des zones colorées selon la classe prédite cmap=plt.cm.Paired : palette de couleurs pour distinguer les classes.
plt.xlabel(iris.feature_names[pair[0]])
plt.ylabel(iris.feature_names[pair[1]])
plt.axis("tight")

# Affichage des points d'apprentissage
for i, color in zip(range(n_classes), plot_colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=color, label=iris.target_names[i], cmap=plt.cm.Paired)
plt.axis("tight")
plt.suptitle("Decision surface of a decision tree using paired features")
plt.legend()
plt.show()
```
